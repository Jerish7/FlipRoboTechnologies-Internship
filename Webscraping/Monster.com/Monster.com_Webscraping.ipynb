# Python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.

```python
#Importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
```

```python
#Connect to web driver
driver=webdriver.Chrome(r"D://chromedriver.exe")  #r converts string to raw string
#If not r, we can use executable_path = "C:/path name"
```

```python
#Getting the website to driver
driver.get('https://www.monsterindia.com/srp/results?query=software%20developer&searchId=f75672d0-0f03-4a74-8607-45bda5e792df')
#When we run this line, automatically the webpage will be opened
```

```python
#Taking the empty lists
job_title=[]
company=[]
location=[]
experience=[]

#As we need to scrap for 5 pages, we are running a for loop
for i in range(0,5):
    for j in driver.find_elements_by_xpath("//h3[@class='medium']/a"):
        job_title.append(j.text)
    for k in driver.find_elements_by_xpath("//span[@class='company-name']/a"):
        company.append(k.text)
    for l in driver.find_elements_by_xpath("//div[@class='col-xxs-12 col-sm-5 text-ellipsis']/span/small"):
        location.append(l.text)
    for e in driver.find_elements_by_xpath("//div[@class='exp col-xxs-12 col-sm-3 text-ellipsis']/span/small"):
        experience.append(e.text)
    driver.find_element_by_xpath("//button[@class='btn-next-prev btn-next']").click()
    time.sleep(3)
```

```python
jobs=pd.DataFrame({})
jobs['Title']=job_title[:117]
jobs['Company Name']=company
jobs['Location']=location[:117]
jobs['Experience']=experience[:117]
jobs
```
